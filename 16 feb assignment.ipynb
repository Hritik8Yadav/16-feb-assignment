```python
#Ans: Database is the collection of information, data in a structured way.
# SQL:- Structured Query Language is the language in which the data is stored in a relational database in a strucuted way
# - The schema of Sql is fixed.
# NoSQL:- Non Structured Query Language is the language in which the data can also be stored in unstructured way.
# - The schema of Sql is not static but it has dynamic schema.
```

```python
# Ans: The DDL or Database Definition Language is the language which is used to define ,modify the database structure.
# CREATE: used for creating a new database
# DROP: used to delete entire database
# ALTER: used to change or modify the database
# TRUNCATE: used to delete all the data from the databse
```

```python
#Ans3: The DML Database Manupulation language is used to add, modify,change, manupulate the data in the databased.
#INSERT: used to add the data in each column.
#UPDATE: used to modify the existing data in the database.
#DELETE: used to delete data from data table.
```
